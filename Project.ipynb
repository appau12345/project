## Histogram
---

```python
plt.hist(nDataset.density, bins= np.arange(0.990,1.003,0.001) , rwidth= 0.7)

plt.title('Densities of Red Wines', fontdict= {'fontweight': 'bold'})
plt.xlabel('Density')
plt.ylabel('Red wine')

plt.show()
```

### From the histogram, most red wines possess densities within the range of  0.996 - 0.997 .
### At the same time few red wines possess densities within the range of 0.990 - 0.991 .

##
## Scatter graph
---

```python
plt.title('Fixed acidity over pH', fontdict= {'fontweight': 'bold'})
plt.scatter(x='pH',y='fixed_acidity',data = nDataset, c= 'r')
plt.xlabel('pH')
plt.ylabel('Fixed acidity')

plt.legend()
```

### From the graph it can be sayd that those with a lower pH had a higher fixed acidity and those with higher pH had a lower fixed acidity.
### Therefore as the pH of a red wine increases, it's fixed acidity decreases.
